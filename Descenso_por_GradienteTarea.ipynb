#### Funcion Elegida
$ f(x) = x^{2}+x*3+x $

### Gráfica de la función y los puntos que se obtienen a medida que se ejecutan los pasos de cálculo (hasta k=3).

```python
import matplotlib.pyplot as plt
import numpy as np
import math 

fun = lambda x : np.power(x, 2) + x * 3 + x
x1 = np.arange(-10,11,0.01)
y1 = fun(x1)
plt.figure(figsize = (10, 5))
plt.axhline(color = 'black')
plt.axvline(color = 'black')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.plot(x1, y1, label = "f(x) = x**2+x*3+x")
plt.scatter(2, fun(2), color = 'red', label = "$P_i$")
plt.grid()
plt.legend()
plt.show()
```

## **Boletin de Practicas 2**

## Gráfica de la función y los puntos que se obtienen a medida que se ejecutan los pasos de cálculo (hasta k=3).

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline

fun = lambda x, y : np.power(x, 2)+2*x*y
x = np.arange(-5.0,6.0,0.1)
y = np.arange(-5.0,6.0,0.1)
X, Y = np.meshgrid(x, y)
Z = fun(X, Y)
fig = plt.figure()
fig.set_size_inches(15,12)
ax = plt.axes(projection = '3d')
ax.plot_surface(X, Y, Z,cmap = 'viridis', edgecolor = 'lightblue')
ax.scatter(3,1,fun(3,1),color='red',label="Punto", s=100, lw=20)
ax.set_title('f(x, y, z)')
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_zlim(-10, 10)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.legend()
plt.show()
```

### Puntos Estacionarios

Para una función diferenciable de varias variables reales , un punto estacionario es un punto en la
superficie del gráfico donde todas sus derivadas parciales son cero (de manera equivalente, el
gradiente es cero)

image.png

Cruces Rojas son Puntos Estacionarios y Puntos Verdes de Inflexión
$ $ \

Los puntos estacionarios son fáciles de visualizar en la gráfica de una función de una variable:
corresponden a los puntos en la gráfica donde la tangente es horizontal (es decir, paralela al eje x ).
Para una función de dos variables, corresponden a los puntos de la gráfica donde el plano tangente
es paralelo al plano xy
